```python
### read a whole file

with open('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
## read a file line by line
with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) ## . strip operation removes nnew line chararcter
```

```python
## writing a file (overwriting)

with open('example.txt','w') as file:
    file.write('hello world\n')
    file.write('this is a new line.')

```

```python
## write a file (without overwriting)
with open('example.txt','a') as file:
    file.write('Append operation is taking place !')
```

```python
## wrting list of lines to a file
lines=['first line\n','second line\n','third line\n']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
## binary files
 
# writing to a binary file
data=b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
    file.write(data)
```

```python
### read a  binary file
with open('example.bin','rb') as file:
    binary_content=file.read()
    print(binary_content)
```

```python
### read the content from a file and write it to another file
## copying content from one file to another
with open('example.txt','r') as source_file:
    content=source_file.read()

with open('copy_example.txt','w') as dest_file:
    dest_file.write(content)

```

```python
##read a text file and count the number of lines, words and characters
with open('example.txt','r') as file:
    content=file.read()
    lines=content.splitlines()
    words=content.split()
    characters=len(content)
    
    print(f'Number of lines: {len(lines)}')
    print(f'Number of words: {len(words)}')
    print(f'Number of characters: {characters}')

```
