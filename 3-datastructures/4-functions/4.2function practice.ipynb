```python
def convert_temprature(temp,unit):
    if unit == 'C':
        return (temp * 9/5) + 32
    elif unit =='F':
        return (temp - 32) * 5/9
    print(convert_temprature(25,'C'))
    print(convert_temprature(77,'F'))
```

```python
def is_strong_password(password):
    if len (password)<8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in '!@#$%^&*()_+-=[]{}|;:,.<>?/' for char in password):
        return False
    return True
```

```python
### calculate the total cost of items ina shopping cart
def calculate_total_cost(cart):
    total_cost=0
    for item in cart:
        total_cost += item['price'] * item['quantity']
## example cart data
cart = [
    {'name':'apple','price':0.6,'quantity':4},
    {'name':'bread','price':2.5,'quantity':2},
    {'name':'milk','price':1.5,'quantity':1}
]

###calling the function
total_cost = calculate_total_cost(cart)
print(total_cost)
```
