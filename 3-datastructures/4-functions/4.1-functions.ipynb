```python
num=2
if num%2==0:
    print("Even")
else:
    print("Odd")
```

```python
def even_odd(num):
    if num%2==0:
        print("Even")
    else:
        print("Odd")
```

```python
even_odd(3)
```

```python
even_odd(2)
```
