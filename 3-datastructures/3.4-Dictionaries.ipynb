```python
### accessing elements
student={"name":"John","age":25,"grade":'A'}
print(student)

```

```python
print(student['name']) ##accessing dictionary elements
print(student['age'])
print(student['grade'])

##acccessing using get method
print(student.get('name'))
print(student.get('age'))
print(student.get('lastname'))
print(student.get('lastname','Not Found'))
## modifying elements
##dictionary are mutable so we can modify them
print(student)
student['age']=26 ##modifying age from 25 to 26
print(student)
student["address"]="downtown"  ##adding new key value pair


```

```python
del student['grade']  ##deleting key value pair
print(student)
```

```python
##dictionary methods
keys=student.keys() ## getting all keys
print(keys)
values=student.values() ## getting all values
print(values)
items=student.items()  ## getting all key value pairs
print(items)

```

```python
##shallow copy 
student_copy=student.copy()
print(student_copy)
print(student)
```

```python
student["name"]="Jane"  ##modifying original dictionary
print(student_copy)  ## change is reflected in copy as well
print(student)
```

```python
##iterating through dictionary
##you can iterate through loops to get keys and values
for keys in student.keys():
    print(keys)
```

```python
## itereate over values
for values in student.values():
    print(values)
```

```python
##iterating over key value pairs
for key,value in student.items():
    print(f"{key}:{value}")
    
```

```python
## nested dictionaries
students={
    "student1":{"name":"John","age":25},
    "student2":{"name":"Jane","age":22},
    "student3":{"name":"Dave","age":23}
}
print(students)
```

```python
### accessing nested dictionary elements
print(students['student1']["name"])
print(students['student1']["age"])
```

```python
##itersating through nested dictionaries
for student_id,student_info in students.items():
    print(f"{student_id}:")
    for key,value in student_info.items():
        print(f"  {key}:{value}")
```

```python
##dictionary comprehension
squares={x:x*x for x in range(5)}
print(squares)
```

```python
##conditional dictionary comprehension
evens={x:x*x for x in range(10) if x%2==0}
print(evens)
```

```python
##practical examples
## use a dict to count the freq of elements in a list
numbers=[1,2,2,3,4,4,4,5]
freq={}
for number in numbers:
    if number in freq:
        freq[number]+=1
    else:
        freq[number]=1
print(freq)
```

```python
##merge two dictionaries
dict1={"a":1,"b":2}
dict2={"b":3,"c":4}
merge_dict={**dict1,**dict2}
print(merge_dict)
```
